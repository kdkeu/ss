### Forward fill (ffill)

Generally the best to prevent data leakage

```python
# Use transform to keep the original index structure
df['Sales'] = df.groupby('Product')['Sales'].ffill()

# Optional: If the VERY first value is NaN, ffill has nothing to grab.
# You might want to backfill (bfill) those specific cases.
df['Sales'] = df.groupby('Product')['Sales'].bfill()
```

### Linear interpolation fill

If you use Linear Interpolation for your training set, be careful! Standard interpolation uses future values to fill past NaNs (it looks ahead to see where the line should end).

While this is fine for historical analysis, if you use it on the very end of your dataset to "prepare" it for a LightGBM prediction, you are essentially giving the model a sneak peek at the future. For real-time machine learning, ffill is almost always the safer bet because it only uses information that was available at that specific moment in time.

```python
df['Sales'] = df.groupby('Product')['Sales'].transform(lambda x: x.interpolate(method='linear'))

# Optional: If the VERY first value is NaN, ffill has nothing to grab.
# You might want to backfill (bfill) those specific cases.
df['Sales'] = df.groupby('Product')['Sales'].bfill()
```

```python
df['Sales'] = df['Sales'].fillna(0)
```

## STL Decomposition
See if there is seasonality in the data.

STL (Seasonal-Trend decomposition using Loess) is a time series decomposition method that separates data into three components: trend, seasonality, and residual (noise). It uses locally weighted regression (Loess) to flexibly estimate smooth patterns, making it robust to non-linear trends and changing seasonal effects. STL is widely used because it handles complex real-world data well and allows users to control the smoothness of each component.

```python
## Make the seasonality plot



# Work on store_1 for the decomposition (fill NaN temporarily for STL)
# s1 = (df_raw[df_raw["store"] == "store_1"]
#       .set_index("date")["sales"]
#       .sort_index()
#       .interpolate("time"))   # temporary linear fill for STL only

df_baguette = df[df['Product'] == 'BAGUETTE'].copy()

stl    = STL(df_baguette['Sales'], period=7, robust=True)
result = stl.fit()

fig, axes = plt.subplots(4, 1, figsize=(14, 10), sharex=True)
for ax, data, label, color in zip(
    axes,
    [df_baguette['Sales'], result.trend, result.seasonal, result.resid],
    ["Observed", "Trend", "Seasonal (weekly)", "Residual"],
    ["steelblue", "darkorange", "seagreen", "slategray"]
):
    ax.plot(data, lw=0.9, color=color)
    ax.set_ylabel(label, fontsize=10)
    ax.xaxis.set_major_formatter(mdates.DateFormatter("%b '%y"))

axes[-1].set_xlabel("Date")
fig.suptitle("STL Decomposition ", fontsize=13)
plt.tight_layout()
plt.show()

# Strength of trend & seasonality
var_resid    = np.var(result.resid)
var_deseason = np.var(result.trend + result.resid)
var_detrend  = np.var(result.seasonal + result.resid)
F_trend      = max(0, 1 - var_resid / var_deseason)
F_seasonal   = max(0, 1 - var_resid / var_detrend)
print(f"Trend strength     : {F_trend:.3f}  (0=none, 1=perfect)")
print(f"Seasonal strength  : {F_seasonal:.3f}  (0=none, 1=perfect)")
```

# Stationarity test

Most classical models (SARIMAX) require stationarity. Even for LightGBM, non-stationarity inflates lag feature importance artificially.
Run two tests:

- ADF (Augmented Dickey-Fuller): null hypothesis is non-stationary. A low p-value (< 0.05) means stationarity. statsmodels.tsa.stattools.adfuller
- KPSS: null hypothesis is stationary. A low p-value means non-stationarity. Use both together — they're complementary.

```python
df_baguette.index = df_baguette['Date']
```

```python
def adf_kpss_report(series, label=""):
    series = series.dropna()
    adf_stat, adf_p, _, _, adf_cv, _ = adfuller(series, autolag="AIC")
    try:
        kpss_stat, kpss_p, _, kpss_cv = kpss(series, regression="c", nlags="auto")
    except Exception:
        kpss_stat, kpss_p = np.nan, np.nan

    print(f"{'─'*50}")
    print(f"{label}")
    print(f"  ADF  stat={adf_stat:.3f}  p={adf_p:.4f}  "
          f"→ {'STATIONARY' if adf_p < 0.05 else 'NON-STATIONARY'}")
    print(f"  KPSS stat={kpss_stat:.3f}  p={kpss_p:.4f}  "
          f"→ {'STATIONARY' if kpss_p > 0.05 else 'NON-STATIONARY'}")


# Raw series
adf_kpss_report(df_baguette['Sales'], "Raw series (baguette)")

# First difference
df_baguette_diff = df_baguette['Sales'].diff().dropna()
adf_kpss_report(df_baguette_diff, "First difference")

# Seasonal difference (period=7)
df_baguette_seas_diff = df_baguette['Sales'].diff(7).dropna()
adf_kpss_report(df_baguette_seas_diff, "Seasonal difference (lag-7)")
```

### Verify I degree (d : differencing order) for ARIMA

d should be the first one for which it looks stationary

```python
# Function to difference the series
def difference(series, d):
    diff_series = series.copy()
    for _ in range(d):
        diff_series = diff_series.diff()
    return diff_series

# Try different d values
d_values = [0, 1, 2]

plt.figure(figsize=(10, 6))

for i, d in enumerate(d_values, 1):
    plt.subplot(len(d_values), 1, i)
    plt.plot(difference(df_baguette['Sales'], d))
    adf_kpss_report(difference(df_baguette['Sales'], d), label=f"d = {d}")
    plt.title(f"d = {d}")
    plt.tight_layout()

plt.show()
```

## ACF and PACF plots
The ACF (autocorrelation function) and PACF (partial autocorrelation function) plots help identify the structure of a time series for models like ARIMA. 

The ACF shows how observations are correlated with past values at different lags, which is useful for detecting the moving average (MA) order \(q\): a sharp cutoff in the ACF suggests the appropriate \(q\). 

The PACF measures the direct correlation between observations at a given lag after removing intermediate effects, helping identify the autoregressive (AR) order \(p\): a sharp cutoff in the PACF indicates the appropriate \(p\). 

If both ACF and PACF decay gradually, it may indicate a mixed ARMA model. Differencing (\(d\)) is determined separately by checking stationarity (e.g., trends or unit roots).

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 4))

plot_acf(df_baguette_diff,  lags=40, ax=axes[0], title="ACF — first-differenced series",  alpha=0.05)
plot_pacf(df_baguette_diff, lags=40, ax=axes[1], title="PACF — first-differenced series", alpha=0.05, method="ywm")

for ax in axes:
    ax.set_xlabel("Lag (days)")
plt.tight_layout()
plt.show()

print("Interpretation guide:")
print("  ACF  : if it cuts off after lag q  → MA(q) component")
print("  PACF : if it cuts off after lag p  → AR(p) component")
print("  Spikes at multiples of 7 → residual weekly seasonality")
```

#  Cross correlation

```python
# 1. Prepare the Data
# Assuming 'df' has columns: ['Date', 'Product', 'Sales']
df['Date'] = pd.to_datetime(df['Date'])

# Pivot so each column is a product's time series
pivot_df = df.pivot(index='Date', columns='Product', values='Sales').fillna(0)

# IMPORTANT: Make data stationary (Differencing)
# This removes trends so we see the relationship between 'changes' in sales
pivot_df_diff = pivot_df.diff().dropna()

target_product = "BAGUETTE"
max_lags = 14  # Looking at the last 2 weeks of influence
results = []

# 2. Calculate CCF for each product vs Target
for col in pivot_df_diff.columns:
    if col == target_product:
        continue
    
    # ccf(x, y) measures how much x leads y
    # We want to see if 'col' leads 'target_product'
    correlation_values = ccf(pivot_df_diff[col], pivot_df_diff[target_product], adjusted=False)
    
    # We only care about positive lags (past of col affecting future of target)
    for lag in range(1, max_lags + 1):
        results.append({
            'Product': col,
            'Lag': lag,
            'Correlation': correlation_values[lag]
        })

# 3. Analyze Results
ccf_results = pd.DataFrame(results)

# Find the most relevant product-lag combinations
top_features = ccf_results.sort_values(by='Correlation', ascending=False).head(50)
print("Top Predictive Lags:")
print(top_features)
```

```python
fig,ax = plt.subplots(figsize=(18,6))

top_features = ccf_results.sort_values(by='Correlation', ascending=False).head(75)

top_features['Label'] = top_features.apply(lambda row: f"{row['Product']} (lag {row['Lag']})", axis=1)
sns.barplot(data=top_features, x='Label', y='Correlation', hue='Lag', ax=ax)
# top_features.plot(kind='bar', x='Product', y='Correlation', figsize=(12,6), title="Top Product-Lag Correlations with BAGUETTE", ax=ax,)

ax.set_xticklabels(ax.get_xticklabels(), rotation=90, ha='right', fontsize=10)

# xlabels = [f"{row['Product']} (lag {row['Lag']})" for _, row in top_features.iterrows()]
# ax.set_xticklabels(xlabels, rotation=90, ha='right', fontsize=10)
```

```python
import pandas as pd

# 1. Pivot to Wide Format
# Rows: Date, Columns: Products
wide_df = df.pivot(index='Date', columns='Product', values='Sales')
wide_df = wide_df.fillna(0)

# 2. Create Lags for all products
# Let's say you want Lags 1, 2, and 3 for every product
lags = [1, 2, 7, 14]
lagged_frames = []

for lag in lags:
    # Shift the entire dataframe and rename columns to 'Product_Lag_N'
    df_lagged = wide_df.shift(lag).add_suffix(f'_Lag{lag}')
    lagged_frames.append(df_lagged)

# Combine all lagged columns into one dataframe
wide_df_lags = pd.concat(lagged_frames, axis=1)

# 3. Clean up
# Lags create NaN values at the start of the series; VIF will fail if these exist
wide_df_lags = wide_df_lags.dropna() 

# Ensure all data is float for the VIF calculation
wide_df_lags = wide_df_lags.astype(float)

wide_df_lags
```

# Feature Engineering

## Just making the lag features w/o split just in case

### Make Lag features, One product

```python
import pandas as pd
import numpy as np

def create_tabular_features(df, target_col, feature, date_col, lags=[1, 2, 3, 7]):
    """
    Converts time series into a supervised learning format.
    """
    # 1. Ensure the data is sorted by ID and Date to prevent shuffling lag values
    df = df.sort_values(by=[feature, date_col]).reset_index(drop=True)
    
    # 2. Create Lags (Target-1, Target-2, etc.)
    for lag in lags:
        df[f'lag_{lag}'] = df.groupby(feature)[target_col].shift(lag)
    
    # 3. Create Rolling Statistics (Moving Averages)
    # Note: We shift(1) first so we don't include the 'current' target in the mean
    for window in [7, 30]:
        df[f'rolling_mean_{window}'] = (
            df.groupby(feature)[target_col]
            .transform(lambda x: x.shift(1).rolling(window=window).mean())
        )
    
    # 4. Extract Date-based features (The 'Time' context LightGBM needs)
    df[date_col] = pd.to_datetime(df[date_col])
    df['day_of_week'] = df[date_col].dt.dayofweek
    df['month'] = df[date_col].dt.month
    df['is_weekend'] = df['day_of_week'].isin([5, 6]).astype(int)
    
    # 5. Drop rows with NaNs created by lagging (the 'cold start' problem)
    return df.dropna().reset_index(drop=True)

# Example Usage:
# df_final = create_tabular_features(your_dataframe, 'sales', 'store_id', 'date')
```

## Split into validation and test sets

```python
# ── Split on the RAW dataframe, before any feature engineering ──────────────
# This prevents lag/rolling features from leaking future values into the past.

FORECAST_HORIZON_DAYS = 30          # hold out last N days as test
CUTOFF_DATE = df["Date"].max() - pd.Timedelta(days=FORECAST_HORIZON_DAYS)

df_validation_raw = df[df["Date"] <= CUTOFF_DATE].copy()
df_test_raw  = df[df["Date"] >  CUTOFF_DATE].copy()

print(f"Validation: {df_validation_raw['Date'].min()} → {df_validation_raw['Date'].max()}  ({len(df_validation_raw):,} rows)")
print(f"Test : {df_test_raw['Date'].min()} → {df_test_raw['Date'].max()}  ({len(df_test_raw):,} rows)")


fig,ax = plt.subplots(figsize=(12,6))
prod = "PAIN AU CHOCOLAT"
mask = df_validation_raw["Product"] == prod
sns.lineplot(data=df_validation_raw.loc[mask], x="Date", y="Sales", ax=ax,label="VALIDATION")
mask = df_test_raw["Product"] == prod
sns.lineplot(data=df_test_raw.loc[mask], x="Date", y="Sales", ax=ax,label="TEST")
ax.set_title(f"Validation/Test Split for {prod}", fontsize=12)

```

```python
# Build features on train only; for test we append it to train so lags look back
# into the training period (avoids cold-start on the test window).

df_combined = pd.concat([df_validation_raw, df_test_raw]).sort_values(["Product", "Date"])

df_featured = create_tabular_features(
    df=df_combined,
    target_col="Sales",
    feature="Product",
    date_col="Date",
    lags=[1, 7, 14],
)
df_featured["Product"] = df_featured["Product"].astype('category').copy()

# Re-split after feature engineering using the same cutoff
train = df_featured[df_featured["Date"] <= CUTOFF_DATE].copy()
test  = df_featured[df_featured["Date"] >  CUTOFF_DATE].copy()

TARGET   = "Sales"
CAT_COLS = ["Product"]
FEATURES = [c for c in train.columns if c not in [TARGET, "Date"]]

X_train, y_train = train[FEATURES], train[TARGET]
X_test,  y_test  = test[FEATURES],  test[TARGET]

print(f"Train shape: {X_train.shape}  |  Test shape: {X_test.shape}")
```

## Hyper parameter tuning with Optuna (with time-series aware split)

### Option 1: Do a split and evaluate the parameters in it

```python

def objective(trial):
    params = {
        "objective":        "regression_l1",   # MAE-optimised
        "metric":           "mae",
        "verbosity":        -1,
        "boosting_type":    "gbdt",
        "n_estimators":     trial.suggest_int("n_estimators", 300, 2000),
        "learning_rate":    trial.suggest_float("learning_rate", 1e-3, 0.3, log=True),
        "num_leaves":       trial.suggest_int("num_leaves", 20, 300),
        "max_depth":        trial.suggest_int("max_depth", 3, 12),
        "min_child_samples":trial.suggest_int("min_child_samples", 5, 100),
        "subsample":        trial.suggest_float("subsample", 0.5, 1.0),
        "colsample_bytree": trial.suggest_float("colsample_bytree", 0.5, 1.0),
        "reg_alpha":        trial.suggest_float("reg_alpha", 1e-4, 10.0, log=True),
        "reg_lambda":       trial.suggest_float("reg_lambda", 1e-4, 10.0, log=True),
    }

    # Time-series CV: last 20 % of training data as internal validation
    val_cutoff = int(len(X_train) * 0.8)
    Xtr, Xval = X_train.iloc[:val_cutoff], X_train.iloc[val_cutoff:]
    ytr, yval = y_train.iloc[:val_cutoff], y_train.iloc[val_cutoff:]

    model = lgb.LGBMRegressor(**params)
    model.fit(
        Xtr, ytr,
        eval_set=[(Xval, yval)],
        callbacks=[lgb.early_stopping(50, verbose=False),
                   lgb.log_evaluation(-1)],
        categorical_feature=CAT_COLS,
    )
    preds = model.predict(Xval)
    return mean_absolute_error(yval, preds)


study = optuna.create_study(direction="minimize",
                             sampler=optuna.samplers.TPESampler(seed=42))
study.optimize(objective, n_trials=50, show_progress_bar=True)

print(f"\nBest MAE (CV): {study.best_value:.4f}")
print("Best params:", study.best_params)
```

### Option 2 (better): Do time series splits and evaluate the parameters in it

```python
# FORECASTING RECURSIVELY (using predicted lags for the test set, which is more realistic)

def recursive_forecast(model, train_df, test_df, features, target, cat_cols,
                        lags, windows):
    """
    Predicts one day at a time, feeding predictions back as lag inputs.
    Avoids using any true future sales values during forecasting.
    """
    # Working history starts as the full training set
    history = train_df.copy()

    predictions = []
    test_dates  = sorted(test_df["Date"].unique())

    for forecast_date in test_dates:
        # Rows we need to predict (one per product on this date)
        day_slice = test_df[test_df["Date"] == forecast_date].copy()

        # ── Rebuild lag & rolling features from history ──────────────────────
        for lag in lags:
            day_slice[f"lag_{lag}"] = day_slice["Product"].map(
                history.groupby("Product")[target].nth(-lag)   # nth from end
            )

        for window in windows:
            day_slice[f"rolling_mean_{window}"] = day_slice["Product"].map(
                history.groupby("Product")[target]
                .apply(lambda x: x.iloc[-window:].mean())
            )

        # ── Predict ──────────────────────────────────────────────────────────
        X_day = day_slice[features]
        day_slice["prediction"] = np.clip(model.predict(X_day), 0, None)
        predictions.append(day_slice[["Product", "Date", target, "prediction"]])

        # ── Append predictions to history so next day's lags are correct ─────
        # We substitute the true Sales with our prediction so no future leakage
        pseudo_row = day_slice.copy()
        pseudo_row[target] = pseudo_row["prediction"]
        history = pd.concat([history, pseudo_row], ignore_index=True)

    return pd.concat(predictions, ignore_index=True)


LAGS    = [1, 7, 14]
WINDOWS = [7, 30]

results = recursive_forecast(
    model       = final_model,
    train_df    = train,          # engineered train set
    test_df     = test,           # engineered test set (features will be recomputed)
    features    = FEATURES,
    target      = TARGET,
    cat_cols    = CAT_COLS,
    lags        = LAGS,
    windows     = WINDOWS,
)

# ── Metrics ──────────────────────────────────────────────────────────────────
mae  = mean_absolute_error(results[TARGET], results["prediction"])
rmse = np.sqrt(mean_squared_error(results[TARGET], results["prediction"]))
mask = results[TARGET] > 0
mape = np.mean(np.abs((results.loc[mask, TARGET] - results.loc[mask, "prediction"])
                       / results.loc[mask, TARGET])) * 100

print(f"{'MAE':<8} {mae:.3f}")
print(f"{'RMSE':<8} {rmse:.3f}")
print(f"{'MAPE':<8} {mape:.2f}%  (non-zero actuals only)")

y_pred_recursive = results["prediction"]
y_pred_recursive
```

Cell 3 — Outlier detection & treatment

```python
# ── Z-score method (flag > 3 SD from mean) ────────────────────────────────────
# Purpose: identify data-entry errors or exceptional events that would
# distort model training. We cap rather than drop to preserve daily continuity.

z_scores = np.abs(stats.zscore(dfb["Sales"]))
outlier_mask = z_scores > 3 # Change this to something larger like 4, here I am keeping it to test the code

print(f"Outliers detected (|z| > 3): {outlier_mask.sum()}")
print(dfb[outlier_mask][["Sales", "Price"]])

# ── IQR method (second opinion) ──────────────────────────────────────────────
Q1, Q3 = dfb["Sales"].quantile([0.25, 0.75])
IQR     = Q3 - Q1
iqr_mask = (dfb["Sales"] < Q1 - 3 * IQR) | (dfb["Sales"] > Q3 + 3 * IQR)
print(f"\nOutliers detected (IQR × 3): {iqr_mask.sum()}")

# ── Visualise ─────────────────────────────────────────────────────────────────
fig, ax = plt.subplots(figsize=(14, 4))
ax.plot(dfb.index, dfb["Sales"], lw=0.8, label="Sales")
ax.scatter(dfb.index[outlier_mask], dfb.loc[outlier_mask, "Sales"],
           color="red", zorder=5, label="Outlier (z>3)")
ax.set_title("Sales with Outliers Flagged"); ax.legend()
plt.tight_layout(); plt.show()

# ── Winsorise: cap at 99th percentile ─────────────────────────────────────────
cap = dfb["Sales"].quantile(0.99)
dfb["Sales_clean"] = dfb["Sales"].clip(upper=cap)
print(f"\nCap applied at 99th pct: {cap:.1f}")
```

Cell 5 — Stationarity tests

```python
# ── ADF & KPSS tests ─────────────────────────────────────────────────────────
# PURPOSE:
#   SARIMAX requires a stationary series (constant mean/variance over time).
#   ADF (Augmented Dickey-Fuller): H0 = unit root (non-stationary).
#     → p < 0.05: reject H0, evidence of stationarity.
#   KPSS: H0 = stationary.
#     → p < 0.05: reject H0, evidence of non-stationarity.
#   Use both together — they test opposite nulls, so agreement gives confidence.
#   If non-stationary → differencing (d parameter in ARIMA) is needed.

def stationarity_report(series, label="Series"):
    print(f"{'─'*50}")
    print(f"  {label}")
    print(f"{'─'*50}")

    adf_stat, adf_p, _, _, adf_cv, _ = adfuller(series.dropna(), autolag="AIC")
    print(f"  ADF  stat={adf_stat:.3f}  p={adf_p:.4f}  "
          f"→ {'STATIONARY ✓' if adf_p < 0.05 else 'NON-STATIONARY ✗'}")
    print(f"       Critical values: { {k: round(v,3) for k,v in adf_cv.items()} }")

    kpss_stat, kpss_p, _, kpss_cv = kpss(series.dropna(), regression="c", nlags="auto")
    print(f"  KPSS stat={kpss_stat:.3f}  p={kpss_p:.4f}  "
          f"→ {'STATIONARY ✓' if kpss_p >= 0.05 else 'NON-STATIONARY ✗'}")
    print()

stationarity_report(dfb["Sales_clean"], "Sales (raw)")

# If non-stationary, check first difference
dfb["Sales_diff1"] = dfb["Sales_clean"].diff()
stationarity_report(dfb["Sales_diff1"].dropna(), "Sales (1st difference)")
```

Cell 6 — Seasonality detection

```python
# ── 1. Periodogram — find dominant cycles ────────────────────────────────────
# PURPOSE: Converts the time series into the frequency domain. Peaks in the
# periodogram indicate repeating cycles. The x-axis shows cycles-per-day,
# so 1/7 ≈ 0.143 = weekly cycle, 1/30 ≈ 0.033 = monthly cycle.

freqs, power = periodogram(dfb["Sales_clean"].fillna(0))
dominant_idx = np.argsort(power)[::-1][:5]

fig, axes = plt.subplots(1, 2, figsize=(14, 4))

axes[0].semilogy(freqs, power, color="steelblue", lw=0.8)
for idx in dominant_idx:
    if freqs[idx] > 0:
        axes[0].axvline(freqs[idx], color="red", alpha=0.5, lw=1,
                        label=f"period≈{1/freqs[idx]:.1f}d")
axes[0].set_xlabel("Frequency (cycles/day)"); axes[0].set_ylabel("Power (log)")
axes[0].set_title("Periodogram"); axes[0].legend(fontsize=8)

print("Top 5 dominant periods (days):")
for idx in dominant_idx:
    if freqs[idx] > 0:
        print(f"  Period ≈ {1/freqs[idx]:.1f} days  (power={power[idx]:.1f})")

# ── 2. STL Decomposition ──────────────────────────────────────────────────────
# PURPOSE: Separates the series into Trend + Seasonal + Residual components.
# Helps confirm the period found above and see if seasonality is stable.
# A large seasonal component relative to residual = strong, modelable seasonality.

PERIOD = 7   # update if periodogram suggests otherwise

stl = STL(dfb["Sales_clean"], period=PERIOD, robust=True)
res = stl.fit()

axes[1].plot(dfb.index, res.seasonal, lw=0.9, color="steelblue")
axes[1].set_title(f"STL Seasonal Component (period={PERIOD})")
axes[1].set_ylabel("Seasonal")
plt.tight_layout(); plt.show()

res.plot()
plt.suptitle("STL Decomposition — Trend / Seasonal / Residual", fontsize=12)
plt.tight_layout(); plt.show()
```

Cell 7 — ACF & PACF

```python
# PURPOSE:
#   ACF (Autocorrelation Function): correlation of the series with its own lags.
#     → Tells us the MA(q) order: cut-off after lag q suggests MA(q).
#     → Spikes at multiples of 7 confirm weekly seasonality.
#   PACF (Partial ACF): correlation at lag k after removing shorter-lag effects.
#     → Tells us the AR(p) order: cut-off after lag p suggests AR(p).
#   Together these guide the (p,d,q)(P,D,Q,s) parameter choices for SARIMAX,
#   though auto_arima will search these automatically in the next cell.

fig, axes = plt.subplots(2, 2, figsize=(14, 8))

plot_acf( dfb["Sales_clean"],         lags=60, ax=axes[0,0], title="ACF — Raw Sales")
plot_pacf(dfb["Sales_clean"],         lags=60, ax=axes[0,1], title="PACF — Raw Sales")
plot_acf( dfb["Sales_diff1"].dropna(),lags=60, ax=axes[1,0], title="ACF — Differenced")
plot_pacf(dfb["Sales_diff1"].dropna(),lags=60, ax=axes[1,1], title="PACF — Differenced")

plt.tight_layout(); plt.show()
```

Cell 9 — Auto SARIMAX fit

```python
# auto_arima performs a stepwise search over (p,d,q)(P,D,Q,s) combinations,
# minimising AIC. We pass Price as an exogenous regressor (X).
# m=7 tells it to consider weekly seasonality.

auto_model = pm.auto_arima(
    y_train,
    X              = X_train,
    m              = 7,              # weekly seasonal period
    seasonal       = True,
    d              = None,           # let it determine differencing order
    D              = None,           # let it determine seasonal differencing
    stepwise       = True,           # faster than exhaustive search
    information_criterion = "aic",
    trace          = True,           # print candidate models
    error_action   = "ignore",
    suppress_warnings = True,
)

print("\n", auto_model.summary())
```

Cell 10 — Residual diagnostics

```python
# PURPOSE: Check that model residuals look like white noise.
#   If patterns remain, the model has missed structure.
#   Ljung-Box test H0 = no autocorrelation in residuals.
#   → p > 0.05 at all lags = residuals are clean.

residuals = pd.Series(auto_model.resid(), index=y_train.index)

fig, axes = plt.subplots(2, 2, figsize=(14, 8))

axes[0,0].plot(residuals, lw=0.8); axes[0,0].set_title("Residuals over time")
axes[0,0].axhline(0, color="red", lw=0.8, linestyle="--")

residuals.hist(bins=40, ax=axes[0,1], color="steelblue", edgecolor="white")
axes[0,1].set_title("Residuals Distribution")

plot_acf(residuals, lags=40, ax=axes[1,0], title="ACF of Residuals")
plot_pacf(residuals, lags=40, ax=axes[1,1], title="PACF of Residuals")

plt.tight_layout(); plt.show()

lb = acorr_ljungbox(residuals, lags=[7, 14, 21, 28], return_df=True)
print("Ljung-Box test (p > 0.05 = residuals are white noise):")
print(lb)
```

Cell 11 — Forecast & evaluate

```python
forecast, conf_int = auto_model.predict(
    n_periods = TEST_DAYS,
    X         = X_test,
    return_conf_int = True,
)
forecast   = np.clip(forecast, 0, None)
forecast_s = pd.Series(forecast,   index=y_test.index)
lower      = pd.Series(np.clip(conf_int[:, 0], 0, None), index=y_test.index)
upper      = pd.Series(conf_int[:, 1],                   index=y_test.index)

# ── Metrics ──────────────────────────────────────────────────────────────────
mae   = mean_absolute_error(y_test, forecast_s)
rmse  = np.sqrt(mean_squared_error(y_test, forecast_s))
mask  = y_test > 0
mape  = np.mean(np.abs((y_test[mask] - forecast_s[mask]) / y_test[mask])) * 100

print(f"{'MAE':<8} {mae:.3f}")
print(f"{'RMSE':<8} {rmse:.3f}")
print(f"{'MAPE':<8} {mape:.2f}%  (non-zero actuals only)")

# ── Plot ──────────────────────────────────────────────────────────────────────
fig, ax = plt.subplots(figsize=(14, 5))

# Show last 60 days of training for context
context = y_train.iloc[-60:]
ax.plot(context.index,    context,      color="steelblue", lw=1, label="Train (context)")
ax.plot(y_test.index,     y_test,       color="black",     lw=1.5, label="Actual")
ax.plot(forecast_s.index, forecast_s,   color="crimson",   lw=1.5, linestyle="--", label="Forecast")
ax.fill_between(forecast_s.index, lower, upper, color="crimson", alpha=0.15, label="95% CI")
ax.axvline(y_test.index[0], color="grey", linestyle=":", lw=1)

ax.set_title(f"SARIMAX Forecast — Baguette\nMAE={mae:.2f}  RMSE={rmse:.2f}  MAPE={mape:.1f}%")
ax.set_ylabel("Sales"); ax.legend()
plt.tight_layout(); plt.show()
```
